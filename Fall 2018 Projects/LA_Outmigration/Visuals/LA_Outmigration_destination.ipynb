# Plotting Out-Migration Destination using PERWT (Person Weight) from 2006 to 2017
PERWT indicates how many persons in the US population are represented by a given person in an IPUMS sample.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/ipums_clean.csv")
```

```python
df.MET2013.unique()
```

```python
df = df[['YEAR', 'MET2013', 'PERWT']]
df.head(5)
```

```python
# for each year, aggregate PERWT per metro area

year_outmigrate_dest = df.groupby(['YEAR', 'MET2013']).agg({'PERWT': 'sum'}).rename(
    columns={'PERWT':'OUT_MIGRATE'}
    )
year_outmigrate_dest.head(5)
```

```python
year_outmigrate_dest.sortlevel(0, ascending=False, inplace=True)    # sort YEAR preparing for plot
```

```python
# get top 5 destinations, sort preparing for plot
dest = year_outmigrate_dest['OUT_MIGRATE'].groupby(level=0, group_keys=False)
top_dest = dest.apply(lambda x: x.order(ascending=True).tail(5))

# drop the year index, only keep the metro index
top_dest = top_dest.reset_index(0, drop=True)
top_dest
```

```python
# Plotting Out-Migration by Top Destination and Year

%matplotlib inline
import matplotlib.pyplot as plt

TOP_DEST = 5
NUM_YEAR = 12

# color assignment
colors = []
color_list = ['r', 'g']
for i in range(NUM_YEAR):
    for j in range(TOP_DEST):
        colors.append(color_list[i%2])

ax = top_dest.plot(kind="barh", figsize=(25, 40), fontsize=25, color=colors)
ax.set_title("LA Out-Migration by Top 5 Destination and Year\n\n", fontsize= 35)
ax.set_ylabel("Metropolitan Area\n", fontsize=30)
ax.set_xlabel("\nNumber of Outmigrators", fontsize=30)
ax.tick_params(axis = 'both', which = 'major', labelsize = 25, labelbottom='on', labeltop='on')

# assign color to y-axis text
i = 0;
for xtick in ax.get_yticklabels():
    xtick.set_color(colors[i])
    i = i+1
    
# create text for year
for i in range(12):
    ax.text(0.9, (i * 1.0/NUM_YEAR) + (1.0/NUM_YEAR * 0.5), 2006+i, transform=ax.transAxes, fontsize=30, color=color_list[i%2])

plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_by_destination_top5.jpg", bbox_inches='tight')
```
