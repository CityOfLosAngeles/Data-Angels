# Plotting Race Characteristics using PERWT (Person Weight) from 2006 to 2017
PERWT indicates how many persons in the US population are represented by a given person in an IPUMS sample.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/ipums_clean.csv")
```

```python
# only keep required columns
df = df[['YEAR', 'RACE', 'PERWT', 'FTOTINC', 'INCTOT', 'MARST', 'OWNERSHPD', 'RENT']]
df.count()
```

```python
# only use data points with valid individual income and when individual income <= family income
out_indiv = df[(df.INCTOT != 9999999) & (df.INCTOT <= df.FTOTINC)]
# only use data points who pay rent
out_indiv = out_indiv[out_indiv.RENT != 0]
out_indiv.count()
```

```python
out_indiv.OWNERSHPD.value_counts()
```

```python
out_indiv.MARST.value_counts()
```

```python
# discard columns no longer needed
out_indiv = out_indiv.drop(['PERWT', 'FTOTINC', 'MARST', 'OWNERSHPD'], axis=1)
out_indiv
```

```python
out_indiv.RENT.median()
```

```python
out_indiv.INCTOT.median()
```

```python
out_indiv.INCTOT = out_indiv.INCTOT/12.0
out_indiv.INCTOT.median()
```

```python
# for each year, for each race, compute its median rent and median income

year_race = out_indiv.groupby(['YEAR', 'RACE']).agg({'RENT':'median', 'INCTOT':'median'}).rename(
    columns={'RENT':'MEDIAN RENT', 'INCTOT':'MEDIAN MONTHLY INCOME'}
    )       
year_race.head()
```

```python
# compute median rent as percentage of median income
year_race['MEDIAN RENT AS PERCENT OF INCOME'] = year_race['MEDIAN RENT']/year_race['MEDIAN MONTHLY INCOME'] * 100.0
year_race.head(60)
```

```python
year_race_rent = year_race['MEDIAN RENT AS PERCENT OF INCOME'].unstack()
year_race_rent.head(50)
```

```python
# Plotting Percent vs Year 

%matplotlib inline
import matplotlib.pyplot as plt

ax = year_race_rent.plot(figsize=(25, 15), legend=True, fontsize=25, linewidth=6)
ax.set_title("Median Rent as % of Median Income (Individual Income) By Race\n", fontsize= 35)
ax.set_xlabel("\nYear", fontsize=0)
# ax.set_ylabel("Percentage (%)\n", fontsize=30)
ax.tick_params(axis = 'both', which = 'major', labelsize = 25)

leg = ax.legend(loc='best', prop={'size': 30});

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(30.0)
    
plt.xlim(left=2005)
plt.ylim(top=400, bottom=0)
plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_median_rent_percentage_income_by_race.jpg", bbox_inches='tight')
```
