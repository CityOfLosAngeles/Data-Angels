# Plotting Race Characteristics using PERWT (Person Weight) from 2006 to 2017
PERWT indicates how many persons in the US population are represented by a given person in an IPUMS sample.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/ipums_clean.csv")
```

```python
df = df[['YEAR', 'RACE', 'PERWT']]
```

```python
# for each year, aggregate PERWT per race
year_race_sum = df.groupby(['YEAR', 'RACE']).agg({'PERWT': 'sum'})

# for each year, compute PERWT percentage per race within that year
year_race_percent = year_race_sum.groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).rename(
    columns={'PERWT':'PERCENT'}
    )

year_race_percent.head()
```

```python
year_race_percent = year_race_percent.unstack()
year_race_percent.head()
```

```python
# Plotting Percent vs Year

%matplotlib inline
import matplotlib.pyplot as plt

ax = year_race_percent.plot(title ="LA Out-Migration by Race", figsize=(15, 10), legend=True, fontsize=15, linewidth=2)
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Percentage", fontsize=15)
leg = ax.legend(["African American", "Asian", "Hispanic", "Mixed", "Others", "White"], loc='best');

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(10.0)
    
plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_by_race.jpg")
```

```python
year_race_sum = year_race_sum.rename(
    columns={'PERWT':'TOTAL'}
    )

year_race_sum = year_race_sum.unstack()
year_race_sum.head()
```

```python
# Plotting Total Number vs Year

ax = year_race_sum.plot(title ="LA Out-Migration by Race", figsize=(15, 10), legend=True, fontsize=15, linewidth=2)
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Total Number", fontsize=15)
leg = ax.legend(["African American", "Asian", "Hispanic", "Mixed", "Others", "White"], loc='best');

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(10.0)

plt.show()
```
