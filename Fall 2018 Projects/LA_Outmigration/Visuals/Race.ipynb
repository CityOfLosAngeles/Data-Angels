# Plotting Race Characteristics using PERWT (Person Weight) from 2006 to 2017
PERWT indicates how many persons in the US population are represented by a given person in an IPUMS sample.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/ipums_clean.csv")
la = pd.read_csv("../data/ipums_la_residents.csv")
```

```python
df = df[['YEAR', 'RACE', 'PERWT']]
```

```python
# for each year, aggregate PERWT per race

year_outmigrate_sum = df.groupby(['YEAR', 'RACE']).agg({'PERWT': 'sum'}).rename(
    columns={'PERWT':'OUT_MIGRATE'}
    )                                                                                  

year_residents_sum = la.groupby(['YEAR', 'RACE']).agg({'PERWT': 'sum'}).rename(
    columns={'PERWT':'RESIDENTS'}
    )                                                                                  
```

```python
combined = pd.concat([year_outmigrate_sum, year_residents_sum], axis=1)

# add the number of out-migrated to residents to give total number of people originally reside in LA
combined['RESIDENTS'] = combined['RESIDENTS'] + combined['OUT_MIGRATE']
```

```python
combined['PERCENT'] = 100.0 * combined['OUT_MIGRATE']/combined['RESIDENTS']
combined.head(36)
```

```python
combined.tail(36)
```

```python
combined_count = combined['OUT_MIGRATE'].unstack()
combined_count.head()
```

```python
combined = combined['PERCENT'].unstack()
combined.head()
```

```python
combined = combined.drop(["Other"], axis=1)
combined.head()
combined_count = combined_count.drop(["Other"], axis=1)
combined_count.head()
```

```python
# Plotting Percent vs Year 

%matplotlib inline
import matplotlib.pyplot as plt

ax = combined.plot(figsize=(25, 15), legend=True, fontsize=25, linewidth=6)
ax.set_title("LA Out-Migration by Race (%)\n", fontsize= 35)
ax.set_xlabel("\nYear", fontsize=30)
ax.set_ylabel("Percentage of Outmigrators\n", fontsize=30)
ax.tick_params(axis = 'both', which = 'major', labelsize = 25)

leg = ax.legend(loc='best', prop={'size': 30});

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(30.0)
    
plt.xlim(left=2005)
plt.ylim(top=6, bottom=1)
plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_by_race.jpg", bbox_inches='tight')
```

```python
# Plotting Out-migration Total vs Year 

%matplotlib inline
import matplotlib.pyplot as plt

ax = combined_count.plot(figsize=(25, 15), legend=True, fontsize=25, linewidth=6)
ax.set_title("LA Out-Migration by Race (Count)\n", fontsize= 35)
ax.set_xlabel("\nYear", fontsize=30)
ax.set_ylabel("Number of Outmigrators\n", fontsize=30)
ax.tick_params(axis = 'both', which = 'major', labelsize = 25)

leg = ax.legend(loc='best', prop={'size': 30});

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(30.0)
      
plt.xlim(left=2005)
plt.ylim(top=250000)
plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_by_race_count.jpg", bbox_inches='tight')
```
