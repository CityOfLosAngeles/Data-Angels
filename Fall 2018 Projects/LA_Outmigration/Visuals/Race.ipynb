# Plotting Race Characteristics using PERWT (Person Weight) from 2006 to 2017
PERWT indicates how many persons in the US population are represented by a given person in an IPUMS sample.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/ipums_clean.csv")
la = pd.read_csv("../data/ipums_la_residents.csv")
```

```python
df = df[['YEAR', 'RACE', 'PERWT']]
```

```python
# for each year, aggregate PERWT per race

year_outmigrate_sum = df.groupby(['YEAR', 'RACE']).agg({'PERWT': 'sum'}).rename(
    columns={'PERWT':'OUT_MIGRATE'}
    )                                                                                  

year_residents_sum = la.groupby(['YEAR', 'RACE']).agg({'PERWT': 'sum'}).rename(
    columns={'PERWT':'RESIDENTS'}
    )                                                                                  
```

```python
combined = pd.concat([year_outmigrate_sum, year_residents_sum], axis=1)
combined.head()
```

```python
combined['PERCENT'] = 100.0 * combined['OUT_MIGRATE']/combined['RESIDENTS']
combined.head()
```

```python
combined = combined['PERCENT'].unstack()
combined.head()
```

```python
# Plotting Percent vs Year 

%matplotlib inline
import matplotlib.pyplot as plt

ax = combined.plot(title ="LA Out-Migration by Race", figsize=(15, 10), legend=True, fontsize=15, linewidth=2)
ax.set_xlabel("Year", fontsize=15)
ax.set_ylabel("Percentage", fontsize=15)
leg = ax.legend(loc='best');

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(10.0)
    
plt.show()

# Save to file
fig = ax.get_figure()
fig.savefig("LA_Out-Migration_by_race.jpg")
```
