```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import re
import csv
import os
plt.style.use('fivethirtyeight')
```

```python
# Change working directory
os.chdir('../Data')
```

```python
df_bigrams=pd.read_csv('bigrams_all_nursing_jobs_cleaned_labeled_manually.csv')
df_bigrams.head()
```

```python
df_hardskills = df_bigrams[df_bigrams['Label']=='Hard skill']
```

```python
len(df_hardskills)
```

```python
df_hardskills.reset_index(inplace=True)
```

```python
df_hardskills.drop(df_hardskills.columns[0],axis=1,inplace=True)
```

```python
df_hardskills.drop(df_hardskills.columns[2],axis=1,inplace=True)
```

```python
df_hardskills.head()
```

```python
##top 50 hardskills
df_hardskills_50 = df_hardskills.head(50)
##top 25 hardskills
df_hardskills_25 = df_hardskills.head(25)
```

```python
word_dict={}
word_dict=pd.Series(df_hardskills.Frequency.values,index=df_hardskills.Bigram).to_dict()
```

```python

wordCloud = WordCloud(max_font_size=60, 
                      background_color='white',
                      width = 1000,
                      height = 1000)
 
wordCloud.generate_from_frequencies(word_dict)
#plt.imshow(wordCloud, interpolation='bilinear')
#plt.axis("off")
plt.figure( figsize=(30,15) )
plt.imshow(wordCloud, interpolation = 'bilinear')
plt.axis('off')
plt.title('Hardskills - All Nursing Jobs')
#plt.figure(figsize=(1000,1000))
#plt.show()
#plt.show()
plt.savefig('All_Nursing_Jobs_Hardskills_frequent_words.png', bbox_inches='tight') 
#wordCloud.to_file("Hardskills_frequent_words.png")
 
```
