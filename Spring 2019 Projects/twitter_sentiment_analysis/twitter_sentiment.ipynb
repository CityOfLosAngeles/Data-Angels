```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import re
import csv
plt.style.use('fivethirtyeight')
```

```python
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
```

```python
df_tweets=pd.read_csv('results.csv', usecols=[9, 10])
```

```python
df_tweets.head()
```

```python
df_negative = df_tweets[df_tweets['pred']==0]
```

```python
df_negative.head()
```

```python
import nltk
from nltk import FreqDist
```

```python
more_stopwords = {'lmaooo', 'oh', 'guess', 'go','one','nyc', 'box', 'doe', 'people', 'say', 'chestnut', 'military'
                 }
STOPWORDS = STOPWORDS.union(more_stopwords)
```

```python
negative_tweet_words = ' '.join(porter.stem(str(cd)) for cd in df_negative['lemmatized'])
```

```python
wordcloud = WordCloud(
                      stopwords=STOPWORDS,
                      max_font_size=60,
                      background_color='white',
                      width = 1000,
                      height = 500
                     ).generate(negative_tweet_words)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud)
plt.axis('off')
#plt.figure(figsize=(1000,1000))
plt.show()
plt.savefig('wordcloud_negative.png', bbox_inches='tight')
```

```python
df_positive = df_tweets[df_tweets['pred']==1]
```

```python
positive_tweet_words = ' '.join(porter.stem(str(cd)) for cd in df_positive['lemmatized'])
```

```python
more_stopwords = {'ha', 'lmndr4rkhq', 'nz', 'dc', 'pic', 'twitter', 'ashamed', 'affordable', 'ark', 'fox', 'qns', 'say', 'stuff',
                  'talk'
                 }
STOPWORDS = STOPWORDS.union(more_stopwords)
```

```python
wordcloud = WordCloud(
                      stopwords=STOPWORDS,
                      max_font_size=60,
                      background_color='white',
                      width = 1000,
                      height = 500
                     ).generate(positive_tweet_words)
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud)
plt.axis('off')
#plt.figure(figsize=(1000,1000))
plt.show()
plt.savefig('wordcloud_positive.png', bbox_inches='tight')
```
